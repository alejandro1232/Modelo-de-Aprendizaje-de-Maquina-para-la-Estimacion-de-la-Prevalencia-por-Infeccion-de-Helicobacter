```python
# Listas para almacenar los resultados
# Itera sobre cada fila del DataFrame
for index, row in dfMap.iterrows():
    lat, lon = row['Latitud'], row['Longitud']
    # Consulta OpenWeatherMap para la temperatura
    weather_url = f"http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={OWM_API_KEY}&units=metric"
    weather_response = requests.get(weather_url)
    weather_data = weather_response.json()
    temperatura = weather_data['main']['temp']
    humedad= weather_data['main']['humidity']
    temperaturas.append(temperatura)
    humedades.append(humedad)


# Añadir los datos al DataFrame original
dfMap['Humedades']= humedades
dfMap['Temperatura'] = temperaturas
```

```python
# Itera sobre cada fila del DataFrame
for index, row in dfMap.iterrows():
   lat, lon = row['Latitud'], row['Longitud']
   elevation_url = f"https://api.open-elevation.com/api/v1/lookup?locations={lat},{lon}"
   elevation_response = requests.get(elevation_url)
   elevation_data = elevation_response.json()
   altitud = elevation_data['results'][0]['elevation']
   altitudes.append(altitud)

# Añadir los datos al DataFrame original
dfMap['Altitud'] = altitudes
```
